# LANGGRAPH

Graph --> Nodes and Edges --> Workflow

#### Langgraph: 

Product or framework of langchain for creating workflows(or pipeline or orchestration)  by creating complex agentic or multi-agentic patterns

From simple AI assistant to complex Agentic pattern(creating DCG) (In Langchain- we create DAG using chains Prompt -> LLM -> Parser -> O/P)

In Langgraph, We have a low level access on patterns

 Graph: Nodes and Edges:

 DAG - Direct Acyclic Graph (No cycles)
 
 DCG - Direct Cyclic Graph  (cycles - loops)  <br> <br>
 


#### Node:
Function (Any custom function) or Tools <br> By default -> startnode(usually i/p) .... endnode(usually o/p)

#### Edges:
Mapping b/w two nodes.
Types:
1. Normal edges
2. Conditional edges

#### State:
Pass the message across to node
